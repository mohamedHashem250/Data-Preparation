# Visualize the effect of Log Scaline Transformation

```python
#from Distrubtion plot:
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
sns.histplot(data = data , x = 'charges',kde = True )
plt.title('Before Log scaling')
plt.subplot(1,2,2)
sns.histplot( x =np.log10(data['charges']),kde = True)
plt.title('After Scaling')
plt.show()#the disturbtion is changed!!!!!!!!!
```

```python
#from BoxPlot View:
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
sns.boxplot(data = data , y = 'charges' )
plt.title('Before scaling')
plt.subplot(1,2,2)
sns.boxplot( y =np.log10(data['charges']))
plt.title('After Scaling')
plt.show()
```

# Visualize The effect of StandardScaler Scaling:

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X = data['charges'].values.reshape(-1,1))
target_scaler  = scaler.transform(X = data['charges'].values.reshape(-1,1))
```

```python
#from Disturbtion plot:
plt.figure(figsize= (14, 4))
plt.subplot(1,2,1)
sns.histplot(data = data, x = 'charges', kde = True)
plt.title('Before StandardScaler')
plt.subplot(1,2,2)
sns.histplot(x = target_scaler.reshape(-1),kde = True)
plt.title("After StandardScaler")
plt.show()#the distrubtion not changes but the range is compressed
```

```python
#from Boxplot chart:
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
sns.boxplot(data = data , y = 'charges' )
plt.title('Before StandardScaler')
plt.subplot(1,2,2)
sns.boxplot( y = target_scaler.reshape(-1))
plt.title('After StandardScaler')
plt.show()#the outlier is not eliminated
```

# Visualize the effect of Min-max transformation

```python
#min max transformation:
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(data['charges'].values.reshape(-1,1))
target_scaler = scaler.transform(data['charges'].values.reshape(-1,1))
```

```python
#from distrubtion plot:
plt.figure(figsize= (14, 4))
plt.subplot(1,2,1)
sns.histplot(data = data, x = 'charges', kde = True)
plt.title('Before MinMax Scaler')
plt.subplot(1,2,2)
sns.histplot(x = target_scaler.reshape(-1),kde = True)
plt.title("After MinMax Scaler")
plt.show() ##the distrubtion not changes but the range is compressed 
```

```python
#from box plot scaling:
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
sns.boxplot(data = data , y = 'charges' )
plt.title('Before MinMax')
plt.subplot(1,2,2)
sns.boxplot( y = target_scaler.reshape(-1))
plt.title('After MinMax')
plt.show()#the outlier is not eliminated
```

# Visualize the effect of Roboust Scaler transformation

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
scaler.fit(X = data['charges'].values.reshape(-1,1))
target_scaler  = scaler.transform(X = data['charges'].values.reshape(-1,1))
```

```python
#from distrubtion plot:
plt.figure(figsize= (14, 4))
plt.subplot(1,2,1)
sns.histplot(data = data, x = 'charges', kde = True)
plt.title('Before Robust Scaler')
plt.subplot(1,2,2)
sns.histplot(x = target_scaler.reshape(-1),kde = True)
plt.title("After Robust Scaler")
plt.show() ##the distrubtion not changes but the range is compressed 
```

```python
#from box plot scaling:
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
sns.boxplot(data = data , y = 'charges' )
plt.title('Before Robust')
plt.subplot(1,2,2)
sns.boxplot( y = target_scaler.reshape(-1))
plt.title('Afte Robust')
plt.show()#the outlier is not eliminated
```
